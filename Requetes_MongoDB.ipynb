### 2. Pour chaque employé, le nombre de clients, le nombre de commandes et le montant total de celles-ci

```python
pandas.DataFrame(list(db.employees.aggregate([
    {"$lookup": { #Jointure entre la table "employees" et "customers" 
        "from": "customers",
        "localField": "employeeNumber",
        "foreignField": "salesRepEmployeeNumber",
        "as": "clients"}},
    {"$unwind": {"path": "$clients", "preserveNullAndEmptyArrays": True}}, #séparation des documents en gardant les NA
    
    {"$lookup": { #Jointure entre la table "orders" et "customers" 
        "from": "orders",
        "localField": "clients.customerNumber",
        "foreignField": "customerNumber",
        "as": "orders"}},
    
    {"$unwind": {"path": "$orders", "preserveNullAndEmptyArrays": True}}, 
    
    {"$unwind": {"path": "$orders.payments", "preserveNullAndEmptyArrays": True}},
    
    {"$group": { #regroupement par numéro d'employé
        "_id": "$employeeNumber",
        "firstname": {"$first": "$firstName"},  
        "lastname": {"$first": "$lastName"}, 
        "nombre_clients": {"$addToSet": "$clients.customerNumber"}, 
        "commandes_uniques": {"$addToSet": "$orders.orderNumber"},
        "montant_total": {"$sum": {"$ifNull": ["$orders.payments.amount", 0]}}}},
    
    {"$project": {
        "firstname": 1,
        "lastname": 1,
        "nombre_clients": {"$size": "$nombre_clients"}, 
        "nombre_commandes": {"$size": "$commandes_uniques"},
        "montant_total": 1}}
])))
```

### 3. Idem pour chaque bureau (nombre de clients, nombre de commandes et montant total), avec en plus le nombre de clients d’un pays différent, s’il y en a ;

```python
pandas.DataFrame(list(db.employees.aggregate([
    {"$lookup": { #Jointure entre la table "employees" et "customers" 
        "from": "customers",
        "localField": "employeeNumber",
        "foreignField": "salesRepEmployeeNumber",
        "as": "clients"}},
    
    {"$unwind": {"path": "$clients", "preserveNullAndEmptyArrays": True}}, 
    
    {"$lookup": {
        "from": "orders",
        "localField": "clients.customerNumber",
        "foreignField": "customerNumber",
        "as": "orders"}},
    
    {"$unwind": {"path": "$orders", "preserveNullAndEmptyArrays": True}},

    {"$unwind": {"path": "$orders.payments", "preserveNullAndEmptyArrays": True}},

    {"$addFields": {
        "client_etranger": {
            "$cond": {
                "if": {"$ne": ["$clients.country", "$office.country"]},
                "then": 1,
                "else": 0}}}},

    {"$group": { #regroupement par code des bureaux
        "_id": "$officeCode",
        "officeCity": {"$first": "$office.city"},  
        "officeCountry": {"$first": "$office.country"}, 
        "nombre_clients": {"$addToSet": "$clients.customerNumber"},
        "nombre_commandes": {"$addToSet": "$orders.orderNumber"},
        "montant_total": {"$sum": {"$ifNull": ["$orders.payments.amount", 0]}},
        "nombre_clients_etrangers": {"$sum": "$client_etranger"}}},

    {"$project": {
        "_id": 0,
        "_id": 1,
        "officeCity" : 1,
        "officeCountry" : 1,
        "nombre_clients": {"$size": "$nombre_clients"},
        "nombre_commandes": {"$size": "$nombre_commandes"},
        "montant_total": 1,
        "nombre_clients_etrangers": 1}}
])))
```

### 5. Donner le nombre de commande pour chaque pays, ainsi que le montant total des commandes et le montant total payé : on veut conserver les clients n’ayant jamais commandé dans le résultat final

```python
pandas.DataFrame(list(db.customers.aggregate([
    {"$lookup": { #jointure entre "customers" et "orders"
            "from": "orders",
            "localField": "customerNumber",
            "foreignField": "customerNumber",
            "as": "customerOrders"}},

    {"$unwind": {"path": "$customerOrders",
                 "preserveNullAndEmptyArrays": True}},
    {"$unwind": {"path": "$customerOrders.order_details",
                 "preserveNullAndEmptyArrays": True}},
    {"$unwind": {"path": "$customerOrders.payments",
                 "preserveNullAndEmptyArrays": True}},
    
    {"$group": {"_id": {"country": "$country","customerNumber": "$customerNumber"}, #regroupement par pays
                "total_payé": {"$sum": "$customerOrders.payments.amount"},
                "total_commande": {"$sum": {"$multiply": ["$customerOrders.order_details.quantityOrdered","$customerOrders.order_details.priceEach"]}},
                "uniqueOrders": {"$addToSet": "$customerOrders.orderNumber"}}},
    
    {"$addFields": {"nombre_commandes": {"$size": "$uniqueOrders"}}},
   
    {"$group": {"_id": "$_id.country",
                "nombre_commandes": {"$sum": "$nombre_commandes"},
                "total_payé": {"$sum": "$total_payé"},
                "total_commande": {"$sum": "$total_commande"}}},

    {"$project": {"_id": 0,
                  "Pays": "$_id",
                  "Nombre_de_commandes": "$nombre_commandes",
                  "Total_Commandé": "$total_commande",
                  "Total_Payé": "$total_payé"}}
])))
```

### 6. On veut la table de contigence du nombre de commande entre la ligne de produits et le pays du client (127 lignes)

```python
pandas.DataFrame(list(db.customers.aggregate([
    {"$lookup": {"from": "orders", #jointure entre "customers" et "orders"
                 "localField": "customerNumber",
                 "foreignField": "customerNumber",
                 "as": "orders"}},

    {"$unwind": "$orders"},

    {"$unwind": "$orders.order_details"},

    {"$lookup": { #jointure entre "products" et "orders" 
            "from": "products",
            "localField": "orders.order_details.productCode",
            "foreignField": "productCode",
            "as": "product_info"}},

    {"$unwind": "$product_info"},

    {"$group": {"_id": {"ligne_produit": "$product_info.productLine", #regroupement par ligne de produit
                        "pays": "$country"},
                "commandes_uniques": {"$addToSet": "$orders.orderNumber"}}},

    {"$project": {"_id": 0,
                  "Ligne_de_Produit": "$_id.ligne_produit",
                  "Pays": "$_id.pays",
                  "Nombre_de_Commandes": {"$size": "$commandes_uniques"}}},

    {"$sort": {"Ligne_de_Produit": 1, "Pays": 1}} #tri par ordre alphabétique
])))
```

### 7. On veut la même table croisant la ligne de produits et le pays du client, mais avec le montant total payé dans chaque cellule 

```python
pandas.DataFrame(list(db.customers.aggregate([

    {"$lookup": {"from": "orders", #jointure entre "customers" et "orders" 
                 "localField": "customerNumber",
                 "foreignField": "customerNumber",
                 "as": "orders"}},

    {"$unwind": {"path": "$orders",
                 "preserveNullAndEmptyArrays": False}},

    {"$lookup": {"from": "products", #jointure entre "products" et "orders"
                 "localField": "orders.order_details.productCode",
                 "foreignField": "productCode",
                 "as": "product_info"}},

    {"$unwind": {"path": "$product_info",
                 "preserveNullAndEmptyArrays": False}},

    {"$unwind": {"path": "$orders.payments",
                 "preserveNullAndEmptyArrays": False}},

    {"$group": {"_id": {"ligne_produit": "$product_info.productLine", #regroupement par ligne de produit
                        "pays": "$country"},
                "montant_total_paye": {"$sum": "$orders.payments.amount"}}},

    {"$project": {"_id": 0,
                  "Ligne_de_Produit": "$_id.ligne_produit",
                  "Pays": "$_id.pays",
                  "Montant_Total_Paye": "$montant_total_paye"}},

    {"$sort": {"Ligne_de_Produit": 1, "Pays": 1}} #tri par ordre alphabétique
])))
```

### 8. Donner les 10 produits pour lesquels la marge moyenne est la plus importante (cf buyPrice et priceEach) ;

```python
pd.DataFrame(list(db.orders.aggregate([

    {"$unwind": {"path": "$order_details",
                 "preserveNullAndEmptyArrays": False}},

    {"$lookup": {"from": "products", #jointure entre "orders" et "products" 
                 "localField": "order_details.productCode",
                 "foreignField": "productCode",
                 "as": "product_info"}},

    {"$unwind": {"path": "$product_info",
                 "preserveNullAndEmptyArrays": False}},

    {"$addFields": {"marge": {"$subtract": ["$order_details.priceEach", "$product_info.buyPrice"]}}}, #ajout de "marge" dans les documents

    {"$group": {"_id": "$order_details.productCode", #regroupement par code produit
                "nom_produit": {"$first": "$product_info.productName"},
                "marge_moyenne": {"$avg": "$marge"}}},

    {"$sort": {"marge_moyenne": -1}}, #tri par ordre décroissant

    {"$limit": 10} #limiter à 10
])))
```

### 9. Lister les produits (avec le nom et le code du client) qui ont été vendus à perte : Si un produit a été dans cette situation plusieurs fois, il doit apparaître plusieurs fois. Une vente à perte arrive quand le prix de vente est inférieur au prix d’achat

```python
pd.DataFrame(list(db.orders.aggregate([
    {"$unwind": {"path": "$order_details",
                 "preserveNullAndEmptyArrays": False}},

    {"$lookup": {"from": "products", #jointure entre "orders" et "products" 
                 "localField": "order_details.productCode",
                 "foreignField": "productCode",
                 "as": "product_info"}},

    {"$unwind": {"path": "$product_info",
                 "preserveNullAndEmptyArrays": False}},
    
    {"$addFields": {"vendu_a_perte": {"$lt": ["$order_details.priceEach", "$product_info.buyPrice"]}}}, #ajout de "vendu_a_perte" dans les documents

    {"$match": {"vendu_a_perte": True}},

    {"$project": {"_id": 0,
                  "code_client": "$customerNumber",
                  "nom_produit": "$product_info.productName",
                  "code_produit": "$order_details.productCode",
                  "prix_vente": "$order_details.priceEach",
                  "prix_achat": "$product_info.buyPrice"}}
])))
```

### 10. (bonus) Lister les clients pour lesquels le montant total payé est supérieur aux montants totals des achats

```python
pandas.DataFrame(list(db.customers.aggregate([

    {"$lookup": {"from": "orders", #jointure entre "customers" et "orders"
                 "localField": "customerNumber",
                 "foreignField": "customerNumber",
                 "as": "orders"}},

    {"$unwind": {"path": "$orders",
                 "preserveNullAndEmptyArrays": True}},

    {"$unwind": {"path": "$orders.order_details",
                 "preserveNullAndEmptyArrays": True}},

    {"$group": {"_id": {"customerNumber": "$customerNumber","customerName": "$customerName"}, #regroupement par numéro client et nom client
                "total_payé": {"$sum": {"$sum": "$orders.payments.amount"}},
                "total_commande": {"$sum": {"$multiply": ["$orders.order_details.quantityOrdered",
                                                        "$orders.order_details.priceEach"]}}}},

    {"$group": {"_id": "$_id.customerNumber", 
                "customerName": {"$first": "$_id.customerName"},
                "total_payé": {"$sum": "$total_payé"},
                "total_commande": {"$sum": "$total_commande"}}},

    {"$match": {"$expr": {"$gt": ["$total_payé", "$total_commande"]}}},

    {"$project": {"_id": 0,
                 "customerNumber": "$_id",
                 "customerName": 1,
                 "total_payé": 1,
                 "total_commande": 1}}
])))
```
