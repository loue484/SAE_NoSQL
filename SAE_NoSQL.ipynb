## Tables Order, Order_details et Payment

**Récupération des données**

```python
orders = pandas.read_sql_query(
    """
    SELECT *
    FROM Orders;
    """,
    conn
)

order_details = pandas.read_sql_query(
    """
    SELECT *
    FROM OrderDetails;
    """,
    conn
)

payments = pandas.read_sql_query(
    """
    SELECT *
    FROM Payments;
    """,
    conn
)
```

**Listes order detail et payments**

```python
liste_order_details = [
    order_details.query('orderNumber == @id')
    .drop(columns = ["orderNumber"])
    .to_dict(orient = "records")
        for id in orders.orderNumber
]
print(liste_order_details)
```

```python
liste_payments = [
    payments.query('customerNumber == @id')
    .drop(columns = ["customerNumber"])
    .to_dict(orient = "records")
        for id in orders.customerNumber
]
print(liste_payments)
```

**Ajout des 2 listes à la table Order**

```python
orders_orderdetails_payments = orders.assign(order_details = liste_order_details, payments = liste_payments)
orders_orderdetails_payments.head()
```

## Tables Customers et Products

**Récupération des données**

```python
customers = pandas.read_sql_query(
    """
    SELECT *
    FROM Customers;
    """,
    conn
)

products = pandas.read_sql_query(
    """
    SELECT *
    FROM Products;
    """,
    conn
)
```
